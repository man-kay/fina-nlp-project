```python
import pandas as pd
import numpy as np
```

```python
#Load train and test result
train_df = pd.read_csv('../../results/bert_finetuned/bert_finetuned_240426_1711_train.csv')
test_df = pd.read_csv('../../results/bert_finetuned/bert_finetuned_240426_1711_test.csv')
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
# Calculate ESG Score MSE
ESG_Score_MSE_train = np.square(np.subtract(train_df['esg_score'],train_df['esg_score_pred'])).mean() 
ESG_Score_MSE_test = np.square(np.subtract(test_df['esg_score'],test_df['esg_score_pred'])).mean()
```

```python
# Calculate accuracy of ESG Risk Level
ESG_Risk_Level_Accuracy_train = (train_df['esg_risk_level'] == train_df['esg_risk_level_pred']).mean()
ESG_Risk_Level_Accuracy_test = (test_df['esg_risk_level'] == test_df['esg_risk_level_pred']).mean()
```

```python
# Results
print('ESG Score MSE (Train):', round(ESG_Score_MSE_train, 5))
print('ESG Risk Level Accuracy (Train):', round(ESG_Risk_Level_Accuracy_train, 5))
print('---------------------------------')
print('ESG Score MSE (Test):', round(ESG_Score_MSE_test, 5))
print('ESG Risk Level Accuracy (Test):', round(ESG_Risk_Level_Accuracy_test, 5))
```
